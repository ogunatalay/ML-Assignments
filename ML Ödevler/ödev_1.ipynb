```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Eğer Jupyter Notebook'ta çalışıyorsanız, bu satırı ekleyin
# %matplotlib inline  # Jupyter Notebook'ta kullanabilirsiniz

# 1. Motosiklet veri kümesini motorverileri.csv dosyasından okuma
try:
    # Farklı ayırıcı denemeleri
    motosiklet_df = pd.read_csv('motorverileri.csv', encoding='utf-8', sep=',')  # Varsayılan ayırıcı: ','
    print("Motosiklet Veri Kümesi:\n", motosiklet_df.head())
except Exception as e:
    print(f"Hata: {e}")

# 2. Sütun adlarını kontrol etme
print("Sütun Adları:", motosiklet_df.columns)

# 3. Motosiklet veri kümesini Excel dosyasına yazma ve tekrar okuma
try:
    motosiklet_df.to_excel('motosikletler.xlsx', index=False)
    okunan_df = pd.read_excel('motosikletler.xlsx')
    print("Excel Dosyasından Okunan Veri:\n", okunan_df)
except Exception as e:
    print(f"Hata: {e}")

# 4. Verileri görselleştirme
try:
    plt.figure(figsize=(10, 6))
    sns.barplot(data=motosiklet_df, x='MARKA', y='FIYAT')
    plt.title('Motosiklet Fiyatları')
    plt.xlabel('Marka')
    plt.ylabel('Fiyat')
    plt.xticks(rotation=45)
    plt.tight_layout()  # Grafiğin düzgün görünmesi için alanı ayarlar
    plt.show()  # Grafiği göster

    # 5. Fiyatların dağılım grafiğini oluşturma
    plt.figure(figsize=(10, 6))
    sns.histplot(motosiklet_df['FIYAT'], bins=10, kde=True)
    plt.title('Motosiklet Fiyatlarının Dağılımı')
    plt.xlabel('Fiyat')
    plt.ylabel('Frekans')
    plt.grid()

    # Grafiği göster
    plt.show()
except Exception as e:
    print(f"Görselleştirme Hatası: {e}")
```
